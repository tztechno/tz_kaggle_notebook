# keroru's AC solution using deque

```python
from collections import deque,defaultdict,Counter
def I():return int(input())

n=I()
s=input()
x=n
q=deque([x])

for i in range(n-1,-1,-1):
    x-=1
    c=s[i]
    if c=='R':
        q.appendleft(x)
    else:
        q.append(x)
    
print(*list(q))
```

# my TLE solution using list

```python
N = int(input())
S = str(input())
 
A=[0]

for i in range(N):
  th=A.index(i)

  if S[i]=='L':
    A.insert(th,i+1)
  else:
    A.insert(th+1,i+1)

print(*A)
```

# my TLE solution using np.array

```python
N = int(input())
S = str(input())
 
import numpy as np
A=np.array([0])
 
for i in range(N):
  th=np.where(A==i)[0][0]
  if S[i]=='L':
    A=np.insert(A,th,i+1,axis=None)
  else:
    A=np.insert(A,th+1,i+1,axis=None)
 
print(*A)
```

# my TLE solution using str

```python
# My TLE solution
N = int(input())
S = str(input())

A='0'

for i in range(N):
    if S[i]=='L':
        A=A.replace(str(i), str(i+1)+' '+str(i))
    elif S[i]=='R':
        A=A.replace(str(i), str(i)+' '+str(i+1))
        
print(A)
```
